```python
import pandas as pd

df = pd.read_csv('winequality-red.csv')

df.info()       # info regarding columns
df.describe()   # info overall descriptive summary
df.shape        # info | rows x columns |
df.columns      # All column names
df['quality'].unique() # returns all unique values
df.isnull().sum() # Checks missing values
df.duplicated()   # True / False based on row
df[df.duplicated()]   # returns dublicate values
df.drop_duplicates(inplace=True) # removes all dublicate
df.corr()   # Describes all Correlations
```
