# EDA With Red Wine Dataset


---

```python
import pandas as pd

df = pd.read_csv('winequality-red.csv')

df.info()       # info regarding columns
df.describe()   # info overall descriptive summary
df.shape        # info | rows x columns |
df.columns      # All column names
df['quality'].unique() # returns all unique values
df.isnull().sum() # Checks missing values
df.duplicated()   # True / False based on row
df[df.duplicated()]   # returns dublicate values
df.drop_duplicates(inplace=True) # removes all dublicate
df.corr()   # Describes all Correlations
```

---
#### Representing correlation on heatmap


```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True)
```

---
#### Visualization of Red Wine Dataset (Quality x Count)

```python
df.quality.value_counts().plot(kind='bar')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

---
#### Visualization depeding on Columns
1. All Overlap
2. Seperate

```python
# 1. Overlapped
for columns in df.columns:
    sns.histplot(df[columns], kde=True)
```

```python
#2. Separate columns
sns.histplot(df['pH'])
```

---
#### Univariate, Bivariate & Multivariate Analysis

```python
sns.pairplot(df)
```

---
#### Categorical Plot

```python
sns.catplot(x='quality', y='alcohol', data=df, kind='box')
```

---
#### Scatter Plot

```python
sns.scatterplot(x='alcohol', y='pH', hue='quality', data=df)
```
