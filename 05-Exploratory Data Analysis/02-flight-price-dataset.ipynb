# EDA On Flight Price Dataset

- This dataset have values based on object. And model only understands numeric value
- So we have to do opr to convert these values into numeric.
---

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
!pip install openpyxl
```

---

# Operation on Column: 'Date_of_journey'

```python
df= pd.read_excel('flight_price.xlsx')

df['date'] = df['Date_of_Journey'].str.split('/').str[0]
df['month'] = df['Date_of_Journey'].str.split('/').str[1]
df['year'] = df['Date_of_Journey'].str.split('/').str[2]

df['date'] = df['date'].astype(int)
df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)

df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head(2)
```

---
# Operation on Column: 'Arrival_Time'

```python
# df= pd.read_excel('flight_price.xlsx')

df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df['arv_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['arv_minute'] = df['Arrival_Time'].str.split(':').str[1]

df['arv_hour'] = df['arv_hour'].astype(int)
df['arv_minute'] = df['arv_minute'].astype(int)

df.drop('Arrival_Time', axis=1, inplace=True)
df.head(4)
```

---
# Operation on Column: 'Dep_Time'

```python
# df= pd.read_excel('flight_price.xlsx')

df['dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['dep_minute'] = df['Dep_Time'].str.split(':').str[1]

df['dep_hour'] = df['dep_hour'].astype(int)
df['dep_minute'] = df['dep_minute'].astype(int)

df.drop('Dep_Time', axis=1, inplace=True)
df.head(4)
```

---
# Operation on Column: 'Total_Stops'

```python
# df= pd.read_excel('flight_price.xlsx')
df['Total_Stops'].unique()
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, np.nan:1})
df.head(2)
```

---
# Operation on Column: 'Route'

```python
df.drop('Route', axis=1, inplace=True)
```

---
# Operation on Column: 'Duration'

```python
df['duration_hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['duration_minute'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]

df['duration_hour'] = df['duration_hour'].fillna(0)
df['duration_minute'] = df['duration_minute'].fillna(0)
df = df[df['duration_hour'] != '5m']

df['duration_hour'] = df['duration_hour'].astype(int)
df['duration_minute'] = df['duration_minute'].astype(int)

df.drop('Duration', axis=1, inplace=True)
df.head(4)
```

---
# Operation on Column: 
- 'Airline, Source, Destination, Additional_Info'

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()

encoded_df = pd.DataFrame(encoder.fit_transform(df[['Airline', 'Source', 'Destination', 'Additional_Info']]).toarray(), columns=encoder.get_feature_names_out())

df = pd.concat([df, encoded_df], axis=1)

df.drop('Airline', axis=1, inplace=True)
df.drop('Source', axis=1, inplace=True)
df.drop('Destination', axis=1, inplace=True)
df.drop('Additional_Info', axis=1, inplace=True)

df.head(2)
```
