# EDA On Flight Price Dataset

- This dataset have values based on object. And model only understands numeric value
- So we have to do opr to convert these values into numeric.
---

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
!pip install openpyxl
```

---

# Operation on Column: 'Date_of_journey'

```python
df= pd.read_excel('flight_price.xlsx')

df['date'] = df['Date_of_Journey'].str.split('/').str[0]
df['month'] = df['Date_of_Journey'].str.split('/').str[1]
df['year'] = df['Date_of_Journey'].str.split('/').str[2]

df['date'] = df['date'].astype(int)
df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)

df.drop('Date_of_Journey', axis=1, inplace=True)
```

# Operation on Column: 'Arrival_Time'

```python
# df= pd.read_excel('flight_price.xlsx')

df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df['arv_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['arv_minute'] = df['Arrival_Time'].str.split(':').str[1]

df['arv_hour'] = df['arv_hour'].astype(int)
df['arv_minute'] = df['arv_minute'].astype(int)

df.drop('Arrival_Time', axis=1, inplace=True)
df.head(4)
```
