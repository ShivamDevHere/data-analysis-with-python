#### Series
1D array like object holds any data type. Ex: Column of a table

---

```python
import pandas as pd

data = [1,2,3,4,5]
series = pd.Series(data)

print(series)
```

```python
# series using list with custom index
data = [1,2,3]
index = ['a', 'b', 'c']

series = pd.Series(data, index=index)

print(series)
```

```python
# Series using Dictionary
data_dict = {'a':1, 'b':2, 'c':3}
series = pd.Series(data_dict)

print(series)
```

#### DataFrame
2D, Mutable, Heterogenous type of data Structure with row and columns

---

```python
# Creating a dataframe from a dictionary of list

data = {
    'Name':['Shvam', 'Soni', 'Kashish'],
    'Age':[12,22,17],
    'City':['Delhi', 'Kailash','Rishikesh']
} 

df = pd.DataFrame(data)

print(df)
```

```python
# Convering a DataFrame into a array

import numpy as np
arr1 = np.array(data)
print(arr1)
```

```python
# Creating a DataFrame from a list of Dictionary

data = [
  {'Name':'Shivam', 'Age':22, 'Smart':'1'},
  {'Name':'Shivam', 'Age':22, 'Smart':'1'}, 
    {'Name':'Shivam', 'Age':22, 'Smart':'1'},  
    {'Name':'Shivam', 'Age':22, 'Smart':'1'}
 ]

df = pd.DataFrame(data)
print(df)
```

```python
# Accessing Elements from DataFrame

df = {
    'Name':['Shvam', 'Soni', 'Kashish'],
    'Age':[12,22,17],
    'City':['Delhi', 'Kailash','Rishikesh']
} 
df = pd.DataFrame(data)
print(df,"\n")

# Accessing Column
print(df['Name'], "\n") # returns Series
print(type(df['Name']),"\n") 

# Accessing Rows
print(df.loc[2])       # finds exact index
print(df.iloc[2])      # fetchs '3' index value no matter its index is even c 

# Accessing Aa Element
print(df.loc[0][0], "\n")
```

```python
# Accessing a Speciified Element using at

df = {
    'Name':['Shvam', 'Soni', 'Kashish'],
    'Age':[12,22,17],
    'City':['Delhi', 'Kailash','Rishikesh']
} 
df = pd.DataFrame(data)

df.at[2,'Name']
df.at[0,'Age']

# Accessing a Speciified Element using iat

df.iat[2,2]

```
