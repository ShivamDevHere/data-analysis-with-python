#### Series
1D array like object holds any data type. Ex: Column of a table

---

```python
import pandas as pd

data = [1,2,3,4,5]
series = pd.Series(data)

print(series)
```

```python
# series using list with custom index
data = [1,2,3]
index = ['a', 'b', 'c']

series = pd.Series(data, index=index)

print(series)
```

```python
# Series using Dictionary
data_dict = {'a':1, 'b':2, 'c':3}
series = pd.Series(data_dict)

print(series)
```

---

#### DataFrame
2D, Mutable, Heterogenous type of data Structure with row and columns

---

```python
# Creating a dataframe from a dictionary of list

data = {
    'Name':['Shvam', 'Soni', 'Kashish'],
    'Age':[12,22,17],
    'City':['Delhi', 'Kailash','Rishikesh']
} 

df = pd.DataFrame(data)

print(df)
```
