# df Manipulation and Analysis with Panda

---

```python
import pandas as pd
```

```python
# Some Display operations
 
df = pd.read_csv('missing-data.csv')

df.head(5)
df.tail(5)
df.describe()  # Statistical Analysis
df.dtypes
```

#### Handling Missing Values 
---

```python
# Viewing Missing val fields

df.isnull()               # True shows Nan present
df.isnull().sum()         # |missing val per column|
df.isnull().any()         # True which column inludes missing val
df.isnull().any(axis=1)   # returns for every row
```

```python
# Replacing all Nan with 0

df_filled = df.fillna(0)
```

```python
# Filling missing values with mean of its column

df['Score_fillNA'] = df['Score'].fillna(df['Score'].mean())
df
```

#### Renaming, Changing Datatypes & Apply functions

---

```python
# Renaming Column

df = df.rename(columns={'Score_fillNA':'Score-FILLED'})
```

```python
# df.dtypes
df['Score-FILLED-INT'] = df['Score-FILLED'].astype(int)
```

```python
# Using function

df['fill_Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['fill_Salary'] = df['fill_Salary'].apply(lambda x:x+2000)
df['fill_Salary'] = df['fill_Salary'].astype(int)
df.head()
```

#### Data Aggregating and Grouping
---

```python
# Group based on Single column City and calc mean of thier Score

grouped_mean = df.groupby('City')['Score'].mean()
print(grouped_mean)
```

```python
# Group based on two column

grouped_sum = df.groupby(['City', 'Age'])['Score'].sum()
print(grouped_sum)
```

```python
# Multiple Aggregate Function Applied

grouped_agg = df.groupby(['Age','City'])['Score'].agg(['mean', 'count', 'sum'])
print(grouped_agg)
```
