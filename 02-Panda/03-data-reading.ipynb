# Data Reading From Different Sources
---

```python
import pandas as pd
from io import StringIO
```

#### Operations with Json 
---

```python
# Reading Json data as DataFrame

data = '[{"emp_id": 101,"name": "Shivam Srivastava","age": 23,"department": "AI/ML","salary": 75000}]'

df = pd.read_json(StringIO(data))
print(df)
```

```python
# DataFrame to Json with index and Records

df.to_json()
df.to_json(orient='index')
df.to_json(orient='records')
```

#### Operation on CSV File
---

```python
pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', header=None)
df.to_csv("wine.csv")
```

#### Reading a table from HTML (website)
- requires to install dependencies
---

```python
url = "https://www.w3schools.com/html/html_tables.asp"
df = pd.read_html(url)
print(df[0])
```

```python
url = "https://www.w3schools.com/html/html_tables.asp"
df = pd.read_html(url, match="Country", header=0)[0]
print(df[0])
```
