```python
# Create a line plot

x = [1,2,3,4,5]
y = [2,4,6,78,0]

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Single Line Plot')
plt.show()
```

```python
# Customized Line Plot
    # linestyle: -, ., --, -.
    # marker: o, *

x = [1,2,3,4,5]
y = [2,4,6,78,0]

plt.plot(x,y, color='green', linestyle='--', marker='o', linewidth=0.5, markersize=7)
plt.grid(True)
```

# Multiple Plot

- `.figure()`: is a frame
- `.subplot(1,2,1)`: is a position for plot
    1. 1: | row |
    2. 2: | column |
    3. 1: position
---

```python
x = [1,2,3,4,5]
y1 = [10,20,70,90,500]
y2 = [5,10,15,20,25]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1, color='red')
plt.title("Plot A")

plt.subplot(2,2,2)
plt.plot(x,y2, color='Green')
plt.title("Plot B")

plt.subplot(2,2,3)
plt.plot(y1,y2, color='yellow')
plt.title("Plot C")

plt.subplot(2,2,4)
plt.plot(y2,x, color='black')
plt.title("Plot D")
```

# Histograms

Represents the distribution of a dataset, They divide data into bins and counts | data point | for every bins

---

```python
data = [1,2,2,3,3,3,3,3,3,4,5,5,5,5,5]

plt.hist(data,bins=5, color='pink',edgecolor='blue')
```
