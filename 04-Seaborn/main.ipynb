---
#### Loading Dataset

- Various CSV Dataset files [Click Me](https://github.com/mwaskom/seaborn-data).)
---

```python
tips = sns.load_dataset('tips')
tips
```

---
#### Scatter Plot
---

```python
# Creating a Scatter plot on loaded dataset

import matplotlib.pyplot as plt

sns.scatterplot(x='tip', y='smoker', data=tips)
plt.title("Tips By Smoker")
plt.show()
```

---
#### Line Plot
---

```python
# Creating a Line Plot

sns.lineplot(x='size', y='tip', data=tips)
```

---
#### Categorical Plot
- Using Bar plot
---

```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar Plot for days and gender')
plt.show()
```

---
#### Boxen Plot
---

```python
sns.boxenplot(x='day', y='total_bill', data=tips)
```

---
#### Box Plot
---

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

---
#### Violin Plot
---

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

---
#### Histogram Plot
- KDE Plot: smoothes the curve
---

```python
sns.histplot(tips['total_bill'], bins=10, kde=True)
```

---
#### KDE Plot
---

```python
sns.kdeplot(tips['total_bill'], fill=True)
```

---
#### Pair Plot
---

```python
sns.pairplot(tips)
```

---
#### HeatMap

- Maps only for datas not string
---

```python
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
