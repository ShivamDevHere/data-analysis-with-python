#### Loading Dataset

- Various CSV Dataset files [Click Me](https://github.com/mwaskom/seaborn-data).)
---

```python
tips = sns.load_dataset('tips')
tips
```

---
#### Scatter Plot
---

```python
# Creating a Scatter plot on loaded dataset

import matplotlib.pyplot as plt

sns.scatterplot(x='tip', y='smoker', data=tips)
plt.title("Tips By Smoker")
plt.show()
```

---
#### Line Plot
---

```python
# Creating a Line Plot

sns.lineplot(x='size', y='tip', data=tips)
```

---
#### Categorical Plot
- Using Bar plot
---

```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar Plot for days and gender')
plt.show()
```

---
#### Box Plot
---

```python
sns.boxenplot(x='day', y='total_bill', data=tips)
```
